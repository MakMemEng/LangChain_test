```python
# cos類似度
def cos_similarity(
    a: gensim.models.keyedvectors.KeyedVectors,
    b: gensim.models.keyedvectors.KeyedVectors):
    return (
        np.dot(a, b) / (
            (np.sqrt(np.dot(a, a))) * (np.sqrt(np.dot(b, b)))
        )
    )
```
